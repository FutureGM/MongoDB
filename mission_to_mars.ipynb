## NASA Mars News

```python
#Latest News Titles and Paragraphs
url = "https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
browser.visit(url)
```

```python
#Iterate through homepage and all articles to find title and paragraph objects

#Append Lists through parsing
news_title_lists = []
url_lists = []
news_url_lists = []
news_paragraph_lists = []

#Set browser as html object
html = browser.html
#Parse HTML using Beautiful Soup
soup = BeautifulSoup(html, "html.parser")
sidebar = soup.find("ul", class_="item_list")
categories = sidebar.find_all("li")
# print(categories)
#Retrieve all elements
for category in categories:
    news_url = category.find("a")["href"]
    url_lists.append(news_url)
    titles = category.find(class_="content_title")
    paragraph = category.find(class_="article_teaser_body")
    news_paragraph_lists.append(paragraph)
    for title in titles:
        headings = title.text.strip()
        news_title_lists.append(title)

#Append news_url_lists by concatenating url with news links

news_url_lists = ["https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest" + url for url in url_lists]
```

## Mars Facts

```python
#Web Scrape Mars Fact Table using Pandas
#Set Url Variable 
url = "https://space-facts.com/mars/"
#Set response variable to requests url retrieval
response = requests.get(url)
```

```python
#Use Beautiful Soup to call and Parse HTML
mars_profile = BeautifulSoup(response.text, "html.parser")
mars_profile
```

```python
#Pull profile resultss from HTML Scrape
results = soup.find('table', class_="tablepress")
eq_dia = results.find("tr", class_="row-1 odd").text
eq_dia.replace("\n", "")
pol_dia = results.find("tr", class_="row-2 even").text
pol_dia.replace("\n", "")
mass = results.find("tr", class_="row-3 odd").text
mass.replace("\n", "")
moons = results.find("tr", class_="row-4 even").text
moons.replace("\n", "")
orbit_dis = results.find("tr", class_="row-5 odd").text
orbit_dis.replace("\n", "")
orbit_per = results.find("tr", class_="row-6 even").text
orbit_per.replace("\n", "")
sur_temp = results.find("tr", class_="row-7 odd").text
sur_temp.replace("\n", "")
first_rec = results.find("tr", class_="row-8 even").text
first_rec.replace("\n", "")
rec_by = results.find("tr", class_="row-9 odd").text
rec_by.replace("\n", "")
```

```python
eq_dia = []
pol_dia = []
mass = []
moons = []
orbit_dis = []
orbit_per = []
sur_temp = []
first_rec = []
rec_by = []
for result in results:
    print(result)
    #Error Handling:
    try:
        eq_dia = result.find("tr", class_="row-1 odd").text
        eq_dia.replace("\n", "")
        pol_dia = result.find("tr", class_="row-2 even").text
        pol_dia.replace("\n", "")
        mass = result.find("tr", class_="row-3 odd").text
        mass.replace("\n", "")
        moons = result.find("tr", class_="row-4 even").text
        moons.replace("\n", "")
        orbit_dis = result.find("tr", class_="row-5 odd").text
        orbit_dis.replace("\n", "")
        orbit_per = result.find("tr", class_="row-6 even").text
        orbit_per.replace("\n", "")
        sur_temp = result.find("tr", class_="row-7 odd").text
        sur_temp.replace("\n", "")
        first_rec = result.find("tr", class_="row-8 even").text
        first_rec.replace("\n", "")
        rec_by = result.find("tr", class_="row-9 odd").text
        rec_by.replace("\n", "")
#         Run to see if available
        if (eq_dia):
            print("______________")
            print(eq_dia)
            print(pol_dia)
            print(mass)
            print(moons)
            print(orbit_dis)
            print(orbit_per)
            print(sur_temp)
            print(first_rec)
            print(rec_by)
            print("______________")
            #Use Pandas to Scrape "Mars Planet Profile" Into Table
            mars_table = {
                #Key: Value
                "Equitorial Diameter": eq_dia,
                "Polar Diameter": pol_dia,
                "Mass": mass,
                "Moons": moons,
                "Orbit Distance": orbit_dis,
                "Orbit Period": orbit_per,
                "Surface Temperature": sur_temp,
                "First Record": first_rec,
                "Recorded By": rec_by
            }
    except Exception as e:
        print(e)
print(mars_table)
```

```python
mars_series = pd.Series(mars_table, index=mars_table.keys())
mars_series.replace("\n","", regex=True)
marsDF = pd.DataFrame(mars_series)
marsDF.columns=["Mars Profile"]
marsfacts_DF = marsDF.replace("\n","", regex=True)
# x.replace("\n","", regex=True)
# x = pd.DataFrame(s.reset_index())
# y = x.drop("index", 1)
# y.replace("\n","", regex=True)
```
